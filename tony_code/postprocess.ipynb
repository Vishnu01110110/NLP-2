```python
import pandas as pd
import pandas as pd
filenames = ["pdp.csv", "pitts_event.csv" ]

# read the csv files into a list of dataframes
dataframes = [pd.read_csv(f) for f in filenames]


# Now you can access the dataframes list to work with the data
print(dataframes[0].columns)
print(dataframes[1].columns)
# rename the columns of dataframes[1] 
dataframes[1] = dataframes[1].rename(columns={'month': 'start_month'})
dataframes[1] = dataframes[1].rename(columns={'day': 'start_day'})
dataframes[1] = dataframes[1].rename(columns={'year': 'start_year'})
dataframes[1] = dataframes[1].rename(columns={'time': 'start_time'})
dataframes[1] = dataframes[1].rename(columns={'venue': 'revenue'})
# remove the 'id' col from dataframes[0]
dataframes[0] = dataframes[0].drop(columns=['id'])
# concatenate the two dataframes, add new col if needed
concatenated_df = pd.concat(dataframes, ignore_index=True)
print(concatenated_df.columns)
# move event_name to the first col
# BEGIN: Move 'event_name' column to the first column
cols = concatenated_df.columns.tolist()
cols = ['event_name'] + [col for col in cols if col != 'event_name']
concatenated_df = concatenated_df[cols]
print(concatenated_df['start_time'])
# END: Move 'event_name' column to the first column

# print(concatenated_df.columns)
# print(concatenated_df.head())
# save the new dataframe to csv 
concatenated_df.to_csv('pitts_combined.csv', index=False)
```

```python
import re
import re
df = pd.read_csv('pittscitypaper_events.csv')
price_column = df['Price']
# create a new column call category
print(price_column)
# Iterate through the price column
for index, value in price_column.items():
    # Check if the value doesn't contain any number or "Free"
    if not (re.search(r'\d', str(value).lower()) or re.search("free", str(value).lower())):
        # Move the value to the category column
        df.loc[index, 'category'] = value
        # Set the price column value to NaN
        df.loc[index, 'Price'] = float('nan')
# save df
df.to_csv('pittscitypaper_events_new.csv', index=False)

# iterate through the description column
description_column = df['Description']
# create a new column called phone_number
df['phone_number'] = ''
for index, value in description_column.items():


    # Match template 1: XXX XXX XXXX
    match_template_1 = re.search(r'\d{3}\s\d{3}\s\d{4}', value)
    

    # Match template 2: (XXX) XXX-XXXX
    match_template_2 = re.search(r'\(\d{3}\) \d{3}-\d{4}', value)
   

    # Match template 3: XXXXXXXXXX
    match_template_3 = re.search(r'\d{10}', value)

    # Match template 4: XXX-XXX-XXXX
    match_template_4 = re.search(r'\d{3}-\d{3}-\d{4}', value)


    if match_template_1:
        df.loc[index, 'phone_number'] = match_template_1[0]
    elif match_template_2:
        df.loc[index, 'phone_number'] = match_template_2[0]
    elif match_template_3:
        df.loc[index, 'phone_number'] = match_template_3[0]
    elif match_template_4:
        df.loc[index, 'phone_number'] = match_template_4[0]
# save df
df.to_csv('pittscitypaper_events_new.csv', index=False)
    
```
