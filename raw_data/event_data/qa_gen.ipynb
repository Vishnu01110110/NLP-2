# CMU EVENTS QA

```python

import csv
# for cmu_events
qa_templates = [
    # 1 Conditioned on Title: Location
    ("Where can I see the \"{Title}\"?", 
     "The event is at {Location}."),

    # 2 Conditioned on Title: Entry
    ("Is there a specific location to enter for the \"{Title}\"?", 
     "Yes, it is held at {Location}."),

    # 3 Conditioned on Title: Duration
    ("When is the \"{Title}\"?", 
     "The event will be open on {Date2} at {Time}."),   


    # 4 Conditioned on Title: Description
    ("Can you give me some information about \"{Title}\"?", 
     "{Description}"),

    
]

# Specify the path to your CSV file
csv_file_path = "cmu_events.csv"



# Specify the path to your CSV file
csv_file_path = "cmu_events.csv"

questions = []
answers = []
# Open the CSV file
with open(csv_file_path, "r") as file:
    # Create a CSV reader object
    csv_reader = csv.DictReader(file)
    
    # Iterate through each row in the CSV file
    for row in csv_reader:
        # Access the values in each row as a dictionary
        # Do something with the row data
        #print(row)
        if row['Date2'] == "Today":
            row['Date2'] = row['Date1']
        for index, template in enumerate(qa_templates):
            q_num = index + 1
            question, answer = template
            if q_num == 1 or q_num == 2:
                if row['Location'] == "":
                    continue
            if q_num == 3:
                if row['Time'] == "" or row['Date2'] == "":
                    continue
                if row['Time'] == "All Day":
                    answer = "The event will be open on {Date2} all day."
            if q_num == 4:
                if row['Description'] == "":
                    continue
            #print(f"Template {index + 1}:")
            #print(question.format(**row))
            #print(answer.format(**row))
            questions.append(question.format(**row))
            answers.append(answer.format(**row))

# Write the questions and answers to a new CSV file
output_csv_file_path = "cmu_events_qa.csv"
with open(output_csv_file_path, "w") as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(["Question", "Answer"])
    for question, answer in zip(questions, answers):
        csv_writer.writerow([question, answer])

```

# PITTS Combined QA

```python
import csv
qa_pairs = [    
    # q1
    ("In which category does the event \"{event_name}\" fall?",
     "The event \"{event_name}\" falls under the category {category}."),
    # q2
    ("Where will the event \"{event_name}\" take place?",
     "The event \"{event_name}\" will take place at {revenue}."),

    # q3 start_time needed
    ("When does the event \"{event_name}\" start?",
     "The event \"{event_name}\" starts on {start_month} {start_day}, {start_year} at {start_time}."),
    
    # q4 end_time needed
    ("When does the event \"{event_name}\" end?",
     "The event \"{event_name}\" ends on {end_month} {end_day}, {end_year} at {end_time}."),
    
    # q5 start and end time needed
    ("How long is the event \"{event_name}\"?",
     "The event \"{event_name}\" starts on {start_month} {start_day}, {start_year} at {start_time} and ends on {end_month} {end_day}, {end_year} at {end_time}."),
    
    # q6 venue needed
    ("Where is the venue for the event \"{event_name}\"?",
     "The venue for the event \"{event_name}\" is {revenue}."),
    
]

# Specify the path to your CSV file
csv_file_path = "pitts_combined.csv"

questions = []
answers = []
# Open the CSV file
with open(csv_file_path, "r") as file:
    # Create a CSV reader object
    csv_reader = csv.DictReader(file)
    
    # Iterate through each row in the CSV file
    for row in csv_reader:
        # change all the column name to lower case
        row = {k.lower(): v for k, v in row.items()}
        #print(row.keys())
        for index, template in enumerate(qa_pairs):
            q_num = index + 1
            question, answer = template
            # remove .0 from all
            row['start_year'] = row['start_year'].split('.')[0]
            row['end_year'] = row['end_year'].split('.')[0]
            row['start_day'] = row['start_day'].split('.')[0]
            row['end_day'] = row['end_day'].split('.')[0]
            
            if q_num == 1 and row['category'] == "":
                continue
            if q_num == 2 and row['revenue'] == "":
                continue
            if q_num == 3 and (row['start_time'] == "" or row['start_month'] == "" or row['start_day'] == "" or row['start_year'] == ""):
                continue
            if q_num == 4 and (row['end_time'] == "" or row['end_month'] == "" or row['end_day'] == "" or row['end_year'] == ""):
                continue
            if q_num == 5 and (row['start_time'] == "" or row['start_month'] == "" or row['start_day'] == "" or row['start_year'] == "" \
                                or row['end_time'] == "" or row['end_month'] == "" or row['end_day'] == "" or row['end_year'] == ""):
                continue
            if q_num == 6 and row['revenue'] == "":
                    continue
            questions.append(question.format(**row))
            answers.append(answer.format(**row))

# Write the questions and answers to a new CSV file
output_csv_file_path = "pitts_combined_qa.csv"
with open(output_csv_file_path, "w") as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(["Question", "Answer"])
    for question, answer in zip(questions, answers):
        csv_writer.writerow([question, answer])
```

# Pittscity Paper

```python

import csv
# for cmu_events
qa_templates = [
    # 1 Conditioned on Title: Location
    ("Where can I see the \"{Title}\"?", 
     "The event is at {Location}."),


    # 3 Conditioned on Title: Time
    ("When is the \"{Title}\"?", 
     "The event will be open on {Time}."),

    
    # 5 Conditioned on Title: Venue
    ("Where is the venue for the event \"{Title}\"?",
     "The venue for the event \"{Title}\" is {Venue}."),

    #6 Region
    ("Which region is the event \"{Title}\" in?",
        "The event \"{Title}\" is in {Region}."),
    #7 category
    ("What category does the event \"{Title}\" fall under?",
        "The event \"{Title}\" falls under the category {Category}."),

]

# Specify the path to your CSV file
csv_file_path = "pittscitypaper_events_new.csv"


questions = []
answers = []
# Open the CSV file
with open(csv_file_path, "r") as file:
    # Create a CSV reader object
    csv_reader = csv.DictReader(file)
    
    # Iterate through each row in the CSV file
    for row in csv_reader:

        # to lowercase
        #row = {k.lower(): v for k, v in row.items()}
        #print(row.keys())
        # rename the column 'category' to 'Category'
        row['Category'] = row['category']
        for index, template in enumerate(qa_templates):
            q_num = index + 1
            question, answer = template
            if q_num == 1 and row['Location'] == "":
                continue
            if q_num == 2 and row['Time'] == "":
                continue
            if q_num == 3 and row['Venue'] == "":
                continue
            if q_num == 4 and row['Region'] == "":
                continue
            if q_num == 5 and row['Category'] == "":
                continue

            questions.append(question.format(**row))
            answers.append(answer.format(**row))

# Write the questions and answers to a new CSV file
output_csv_file_path = "pittscity_qa.csv"
with open(output_csv_file_path, "w") as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(["Question", "Answer"])
    for question, answer in zip(questions, answers):
        csv_writer.writerow([question, answer])

```

# csv to txt

```python
# read three csv files
import csv
files = ["cmu_events_qa.csv", "pitts_combined_qa.csv", "pittscity_qa.csv"]
questions = []
answers = []
# Open the files and load questions and answers

for file in files:
    with open(file, "r") as f:
        csv_reader = csv.reader(f)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            question, answer = row
            questions.append(question)
            answers.append(answer)

# Print questions and answers in Q: A: format
output_file_path = "questions_and_answers.txt"
with open(output_file_path, "w") as f:
    for question, answer in zip(questions, answers):
        f.write(f"Q: {question}\n")
        f.write(f"A: {answer}\n")
```
