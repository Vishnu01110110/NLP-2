```python
import random
filenames = ['event_qa.txt', 'test_set_general_info_llama.txt', 'test_set_sports_llama.txt', 'QA_pairs_Music_and_Culture.txt']
# read qa_event.txt
with open(filenames[0], 'r') as f:
    lines = f.readlines()
    questions_event = []
    answers_event = []

    for line in lines:
        if line.startswith('Q:'):
            question = line.strip().replace('Q:', '').strip()
            questions_event.append(question)
        elif line.startswith('A:'):
            answer = line.strip().replace('A:', '').strip()
            answers_event.append(answer)

# read test_set_general_info_llama.txt
with open(filenames[1], 'r') as f:
    lines = f.readlines()
    questions_general_info = []
    answers_general_info = []

    for line in lines:
        if line.startswith('Q:'):
            question = line.strip().replace('Q: Q: ', '').strip()
            questions_general_info.append(question)
        elif line.startswith('A:'):
            answer = line.strip().replace('A: A: ', '').strip()
            answers_general_info.append(answer)

# read test_set_sports_llama.txt
with open(filenames[2], 'r') as f:
    lines = f.readlines()
    questions_sports = []
    answers_sports = []

    for line in lines:
        if line.startswith('Q:'):
            question = line.strip().replace('Q: Q: ', '').strip()
            questions_sports.append(question)
        elif line.startswith('A:'):
            answer = line.strip().replace('A: A: ', '').strip()
            answers_sports.append(answer)
with open(filenames[3], 'r') as f:
    lines = f.readlines()
    questions_music_culture = []
    answers_music_culture = []

    for line in lines:
        if line.startswith('Q:'):
            question = line.strip().replace('Q:', '').strip()
            questions_music_culture.append(question)
        elif line.startswith('A:'):
            answer = line.strip().replace('A:', '').strip()
            answers_music_culture.append(answer)
# take 100 from each and write to a new file
# Combine questions and answers into pairs
qa_event = list(zip(questions_event, answers_event))
qa_general_info = list(zip(questions_general_info, answers_general_info))
qa_sports = list(zip(questions_sports, answers_sports))
qa_music_culture = list(zip(questions_music_culture, answers_music_culture))

# Shuffle the pairs
random.seed(0)
random.shuffle(qa_event)
random.shuffle(qa_general_info)
random.shuffle(qa_sports)
random.shuffle(qa_music_culture)

# Separate the shuffled pairs back into questions and answers
questions_event, answers_event = zip(*qa_event)
questions_general_info, answers_general_info = zip(*qa_general_info)
questions_sports, answers_sports = zip(*qa_sports)
questions_music_culture, answers_music_culture = zip(*qa_music_culture)

questions_test = questions_event[:100] + questions_general_info[:100] + questions_sports[:100] + questions_music_culture[:100]
answers_test = answers_event[:100] + answers_general_info[:100] + answers_sports[:100] + answers_music_culture[:100]
# write to two files for training and testing and for questoin and answer
with open('questions.txt', 'w') as f:
    for question in questions_test:
        f.write(question + '\n')
with open('reference_answers.txt', 'w') as f:
    for answer in answers_test:
        f.write(answer + '\n')
questions_train = questions_event[100:] + questions_general_info[100:] + questions_sports[100:] + questions_music_culture[100:]
answers_train = answers_event[100:] + answers_general_info[100:] + answers_sports[100:] + answers_music_culture[100:]
with open('questions_train.txt', 'w') as f:
    for question in questions_train:
        f.write(question + '\n')
with open('reference_answers_train.txt', 'w') as f:
    for answer in answers_train:
        f.write(answer + '\n')


```
