```python
import random
filenames = ['qa_event.txt', 'test_set_general_info_llama.txt', 'test_set_sports_llama.txt']
question_file = 'questions.txt'
answer_file = 'reference_answers.txt'

questions = []
answers = []
# read qa_event.txt
with open(filenames[0], 'r') as f:
    lines = f.readlines()
    questions = []
    answers = []

    for line in lines:
        if line.startswith('Q:'):
            question = line.strip().replace('Q:', '')
            questions.append(question)
        elif line.startswith('A:'):
            answer = line.strip().replace('A:', '')
            answers.append(answer)
# only take random 700 questions and answers

# remove duplicates
questions = list(set(questions))
answers = list(set(answers))
length = len(questions)
random.seed(0)
random_indices = random.sample(range(length), 750)
questions = [questions[i] for i in random_indices]
answers = [answers[i] for i in random_indices]


# read test_set_general_info_llama.txt
with open(filenames[1], 'r') as f:
    lines = f.readlines()

    for line in lines:
        if line.startswith('Q:'):
            question = line.strip().replace('Q: Q: ', '')
            questions.append(question)
        elif line.startswith('A:'):
            answer = line.strip().replace('A: A: ', '')
            answers.append(answer)

# read test_set_sports_llama.txt
with open(filenames[2], 'r') as f:
    lines = f.readlines()

    for line in lines:
        if line.startswith('Q:'):
            question = line.strip().replace('Q: Q: ', '')
            questions.append(question)
        elif line.startswith('A:'):
            answer = line.strip().replace('A: A: ', '')
            answers.append(answer)
print("Questions:")
print(questions)
print("Answers:")
print(answers)
print("Number of questions: ", len(questions))
print("Number of answers: ", len(answers))
with open(question_file, 'w') as f:
    for question in questions:
        # strip the leading and trailing whitespaces
        question = question.strip()
        f.write(question + '\n')
with open(answer_file, 'w') as f:
    for answer in answers:
        answer = answer.strip()
        f.write(answer + '\n')
```
